## Import Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## T0 vs T4

```python
t0_t4_mscn = pd.read_excel('/content/drive/MyDrive/BM5020/Results/T0 vs T4/T0_vs_T4_ablation_mscn_mod_3_classification_report_all_fold.xlsx')
t0_t4_mscn_se = pd.read_excel('/content/drive/MyDrive/BM5020/Results/T0 vs T4/T0_vs_T4_ablation_mscn_se_mod_3_classification_report_all_fold.xlsx')
t0_t4_full = pd.read_excel('/content/drive/MyDrive/BM5020/Results/T0 vs T4/T0_vs_T4_fusion_early_mod_3_classification_report_all_fold.xlsx')
```

```python
t0_t4_mscn_accuracy = round(t0_t4_mscn['accuracy'].values[0],3)
t0_t4_mscn_precision = round(t0_t4_mscn['macro avg'].values[0],3)
t0_t4_mscn_recall = round(t0_t4_mscn['macro avg'].values[1],3)
t0_t4_mscn_mf1 = round(t0_t4_mscn['macro avg'].values[2],3)
t0_t4_mscn_cohen = round(t0_t4_mscn['cohen'].values[0],3)

t0_t4_mscn_se_accuracy = round(t0_t4_mscn_se['accuracy'].values[0],3)
t0_t4_mscn_se_precision = round(t0_t4_mscn_se['macro avg'].values[0],3)
t0_t4_mscn_se_recall = round(t0_t4_mscn_se['macro avg'].values[1],3)
t0_t4_mscn_se_mf1 = round(t0_t4_mscn_se['macro avg'].values[2],3)
t0_t4_mscn_se_cohen = round(t0_t4_mscn_se['cohen'].values[0],3)

t0_t4_full_accuracy = round(t0_t4_full['accuracy'].values[0],3)
t0_t4_full_precision = round(t0_t4_full['macro avg'].values[0],3)
t0_t4_full_recall = round(t0_t4_full['macro avg'].values[1],3)
t0_t4_full_mf1 = round(t0_t4_full['macro avg'].values[2],3)
t0_t4_full_cohen = round(t0_t4_full['cohen'].values[0],3)
```

```python
# set width of bar
barWidth = 0.1
fig = plt.subplots(figsize =(12, 8))

# set height of bar
mscn_scores = [t0_t4_mscn_accuracy,t0_t4_mscn_precision,t0_t4_mscn_recall,t0_t4_mscn_mf1,t0_t4_mscn_cohen]
print('MSCN:',mscn_scores)
mscn_se_scores = [t0_t4_mscn_se_accuracy,t0_t4_mscn_se_precision,t0_t4_mscn_se_recall,t0_t4_mscn_se_mf1,t0_t4_mscn_se_cohen]
print('MSCN_SE:',mscn_se_scores)
full_scores = [t0_t4_full_accuracy,t0_t4_full_precision,t0_t4_full_recall,t0_t4_full_mf1,t0_t4_full_cohen]
print('FULL:',full_scores)

# Set position of bar on X axis
br1 = np.arange(len(mscn_scores))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, mscn_scores, color ='r', width = barWidth, edgecolor ='grey', label ='MSCN')
plt.bar(br2, mscn_se_scores, color ='g', width = barWidth, edgecolor ='grey', label ='MSCN_SE')
plt.bar(br3, full_scores, color ='b', width = barWidth, edgecolor ='grey', label ='FULL')

# Adding Xticks
plt.xlabel('T0 vs T4', fontweight ='bold', fontsize = 18)
plt.ylabel('Evaluation Score', fontsize = 18)
plt.xticks([r + barWidth for r in range(len(mscn_scores))], ['Accuracy', 'Precision', 'Recall', 'MF$_{1}$', 'Cohen Kappa'], fontsize = 18)

plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),ncol=3, fancybox=True, shadow=True)
plt.show()
```
